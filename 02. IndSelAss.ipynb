# Indexamiento por etiqueta
El valor del índice o el nombre de la columna son sus etiquetas. No necesariamente coincide el valor del índice con la posición, menos la etiqueta de la columna con su posición.

Para consultar los datos por etiquetas, use .loc.

```python
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

# Selección condicional
Esto permitirá consultar datos del Dataframe en base a condiciones que se cumplan en las columnas del mismo

```python
reviews.country == 'Italy'
```

```python
reviews.loc[reviews.country == 'Italy']
```

```python
reviews[reviews.country == 'Italy']
```

```python
(reviews.country == 'Italy') & (reviews.points >= 90)
```

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
reviews[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

```python
reviews[(reviews.country!='Italy')]
```

```python
reviews[(reviews.country!='Italy')]
```

```python
reviews.country.isin(['Italy', 'France'])
```

```python
reviews.loc[reviews.price.notnull()]
```

```python
reviews.loc[reviews.price.isnull()]
```
